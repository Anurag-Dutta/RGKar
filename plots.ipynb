```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True


data = pd.read_csv('us.csv')

years = data['Year']
cases = data['Cases']


plt.figure(figsize=(15, 9))
plt.bar(years, cases, color='white', linewidth=1, edgecolor='blue')
plt.bar(years, cases, color='blue', alpha=0.2)

plt.xlabel('Year', labelpad=30, fontsize=30)
plt.ylabel('Rape cases in United States', labelpad=30, fontsize=30)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)


plt.savefig('plot_us.pdf', format='pdf', bbox_inches='tight')

plt.show()
```
